```python
def text_normalization(text):
    # 소문자화
    text = text.lower()
    # URL 대체
    text = re.sub(r'http\S+|www.\S+', '<URL>', text)
    # 전화번호 대체
    text = re.sub(r'\d{2,3}-\d{3,4}-\d{4}', '<PHONE>', text)
    # 특수 문자 및 구두점 제거
    text = re.sub(r'^\w\s', '', text)
    # 여러 개의 공백을 하나로 수정
    text = re.sub(r'\s+', ' ', text)
    # 반복 문자 처리
    pattern = re.compile(r'(.)\1{2,}')
    text = pattern.sub(r'\1\1', text)
    # 약어 및 은어 처리
    abbr = {
        "취준": "취업 준비"
    }
    for k, v in abbr.items():
        text = text.replace(k, v)
    return text
```

```python
text = """
Good Morning!!! 저는 홍길동입니다. 제 이메일은 hong@gildong.com 이고, 홈페이지는 http://gildong.com 이며, 전화번호는 010-1234-5678 입니다.
요즘 취준 때문에 걱정이 많네요... 다들 건강 조심하세요ㅎㅎㅎ!!!
"""
print(text_normalization(text))
```

```python
import re

text = "Hello, world! How are you doing today? This is an example of sentence \
tokenization. Let`s see how it works. BTW, Dr. Kim has not medical doctor."

#정규식을 이용한 문장 토큰화
sentence_endings = re.compile(r'[.!?]')
sentences = sentence_endings.split(text)

#공백 제거
sentences = [sentence.strip() for sentence in sentences if sentence.strip() != '']

print("문장 토큰화 결과:")
for idx, sentence in enumerate(sentences):
    print(f"문장 {idx + 1}: {sentence}")
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
text = "Hello, world! How are you doing today? This is an example of sentence \
tokenization. Let`s see how it works. BTW, Dr. Kim has not medical doctor."
doc = nlp(text)
sentences = list(doc.sents)
print("문장 토큰화 결과:")
for idx, sentence in enumerate(sentences):
    print(f"문장 {idx + 1}: {sentence}")
```

```python
import re

text = "Hello! How are you doing today? I`m doing fine."
# findall: 문자열에서 정규 표현식 패턴에 맞는 모든 부분을 찾아서 리스트로 반환
# \b 단어 경계를 의미, 공백이나 문장 부호로 판단
# \w 알파벳, 숫자, 밑줄을 의미
# [\w`] 대괄호 안에 포함된 준자 중 하나가 있을 수 있다는 의미. 여기서 `는 따옴표가 아님
# * 바로 앞의 문자가 0번 이상 반복될 수 있음을 의미
tokens = re.findall(r'\b\w[\w`]*\b', text)
print(tokens)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#예제 텍스트
text = "This is a sample sentence, showing off the stop wotds filtration."
#토큰화
words = word_tokenize(text)
print("원본 단어들: ", words)
#불용어 목록
stop_words = set(stopwords.words('english'))
print("불용어 목록: ", list(stop_words)[:10]) # 일부만 출력
#불용어 제거
filtered_words = [word for word in words if word.lower() not in stop_words]
print("불용어 제거 단어들: ", filtered_words)
```
